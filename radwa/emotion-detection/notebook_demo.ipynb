```python
from fer import FER
import cv2
import numpy as np

img = cv2.imread(r"E:\Program Files\Downloads\emotions\Surprised_17.jpg")

# ensure 3 channels (FER expects RGB)
if img.ndim == 2:
    img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
elif img.shape[2] == 4:  # if image has alpha channel
    img = cv2.cvtColor(img, cv2.COLOR_BGRA2BGR)

detector = FER(mtcnn=True)
emotion, score = detector.top_emotion(img)

print(emotion, score)
```

```python
from deepface import DeepFace

result = DeepFace.analyze(img_path=r"E:\Program Files\Downloads\emotions\Surprised_17.jpg", actions=['emotion'])
dominant_emotion = result[0]['dominant_emotion']
confidence = result[0]['emotion'][dominant_emotion]
print(dominant_emotion, confidence)
```
